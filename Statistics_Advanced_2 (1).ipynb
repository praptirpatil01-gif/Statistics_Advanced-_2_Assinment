Q1.What is hypothesis testing in statistics?

Ans:-

Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data.

Explanation:

* It involves formulating two competing hypotheses:

    * Null hypothesis (H₀)

    * Alternative hypothesis (H₁ or Hₐ)

* Sample data is analyzed to determine whether there is enough evidence to reject the null hypothesis.

* The decision is based on probability and significance level.

Example:

A company claims the average weight of packets is 50 g. A sample is tested to check whether this claim is true.

---
Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans:-

**Null Hypothesis (H₀):**

* States that there is no effect, no difference, or no change.

* It is the default assumption.

Example:

H₀: The average weight of packets is 50 g.

**Alternative Hypothesis (H₁):**

* States that there is an effect or a difference.

* It contradicts the null hypothesis.

Example:
H₁: The average weight of packets is not 50 g.

---
Q3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Ans:-

**Significance Level (α):**

* It is the probability of rejecting the null hypothesis when it is actually true.

* Common values: 0.05, 0.01

**Role in Decision Making:**

* If p-value ≤ α → Reject H₀

* If p-value > α → Fail to reject H₀

Example:

At α = 0.05, there is a 5% risk of making a wrong decision by rejecting a true null hypothesis

---
Q4.What are Type I and Type II errors? Give examples of each.

Ans:-

**Type I Error:**

* Rejecting a true null hypothesis.

* Probability = α

Example:

Rejecting a medicine as unsafe when it is actually safe.

**Type II Error:**

* Failing to reject a false null hypothesis.

* Probability = β

Example:

Accepting a faulty machine as working correctly.

Q5.What is the difference between a Z-test and a T-test? Explain when to use each.

Ans:-

**Z-Test:**

* Population standard deviation is known

* Sample size n ≥ 30

* Data is normally distributed

**T-Test:**

* Population standard deviation is unknown

* Sample size n < 30

* Uses sample standard deviation

Q6.Python program to generate a binomial distribution (n=10, p=0.5) and plot its histogram.

Ans:-

Explanation:

* The histogram shows the frequency of successes from 0 to 10.

* The distribution is symmetric around 5.

```python
import random
import matplotlib.pyplot as plt

# Generate binomial data
data = [sum(1 for _ in range(10) if random.random() < 0.5) for _ in range(1000)]

# Plot histogram
plt.hist(data)
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

Q7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


Interpretation:

* If p-value > 0.05, we fail to reject H₀

* The sample mean does not significantly differ from 50

```python
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_mean = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
```

Q8.Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

Explanation:

* The confidence interval gives the range in which the true mean lies with 95% certainty.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(50, 5, 1000)

mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

confidence_interval = stats.norm.interval(0.95, mean, std/np.sqrt(n))

print("95% Confidence Interval:", confidence_interval)

plt.hist(data)
plt.title("Normal Distribution")
plt.show()
```

Q9.Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

Explanation:

* Z-scores represent how many standard deviations a value is away from the mean.

* Positive Z-score → above mean

* Negative Z-score → below mean

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return [(x - mean) / std for x in data]

data = np.random.normal(100, 10, 1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores)
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.show()
```
